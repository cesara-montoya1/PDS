# Ejemplo personal. Muestreo de señales analógicas.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt

# Variables globales
pi = np.pi
```

```python
# Señal a muestrear
x = lambda t: np.sin(2*pi*t) + 2*np.sin(2*(2*pi)*t)

# Señal pseudoanalógica
t_analogical = np.arange(0, pi/2, 1e-3)
x_analogical = x(t_analogical)

# Señal digital
fs = 20
t_digital = np.arange(0, pi/2, 1/fs)
x_digital = x(t_digital)
```

Para la demostración del muestreo de una señal analógica se hará uso de una señal sinusoidal pseudoanalógica (no es realmente analógica porque para esto debería ser continua para todo valor de $t$, cosa que es imposible almacenar en sistemas digitales como un computador), y para la señal digital se muestreará la misma señal con una frecuencia de muestreo ($f_s$) visiblemente menor.

```python
# Grafica de la señal analógica y digital
plt.figure()

plt.plot(t_analogical, x_analogical, 'b', label = "Señal analógica")
plt.stem(t_digital, x_digital, 'r', label = "Señal digitalizada")

plt.xlabel("Tiempo")
plt.ylabel("Amplitud")

plt.grid()
plt.legend()
plt.show()
```
