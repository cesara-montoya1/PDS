# Ejemplo 2. Densidad Espectral de Energia.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 20
u_n = np.arange(N)  # Hace referencia a u(n) = heaviside(n), o sea, solo existe despues de 0
a = 0.5             # Parametro variable
x = a**u_n

plt.figure()
plt.stem(x)
plt.ylabel('Amplitud')
plt.xlabel('Numero de muestras')
plt.grid()
plt.show()
```

```python
def fourierTransform(x_n, alpha):
    w = np.arange(-np.pi, np.pi, alpha)
    senalF = np.zeros(w.size, 'complex_')
    for k in np.arange(x_n.size):
        senalF += x_n[k] * np.exp(-1j*w*k) # Expresion para calcular la transformada
    return senalF * 0.707
```

```python
alpha = 0.01 # Tama√±o de paso
Sxx = np.abs(fourierTransform(x, alpha))**2
w = np.arange(-np.pi, np.pi, alpha)

plt.figure()
plt.plot(w, Sxx)
plt.show()
```

Ahora se va a hacer con $a=-0.5$

```python
a = -0.5             # Parametro variable
x2 = a**u_n

Sxx = np.abs(fourierTransform(x2, alpha))**2
w = np.arange(-np.pi, np.pi, alpha)

plt.figure()
plt.plot(w, Sxx)
plt.show()
```
