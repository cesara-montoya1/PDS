# Ejemplo 2. Transformada de Fourier.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fourier
```

### Transformada de Fourier de una señal $sinc(\omega t)$

```python
t = np.arange(-100, 100, 0.07)
x = np.sinc(2*np.pi*0.2*t)
```

```python
plt.figure()
plt.title('Señal $x(n)$', fontsize=18)
plt.stem(t, x)
plt.xlabel('n')
plt.xlim([-4, 4])
plt.grid()
plt.show()
```

```python
X = fourier.fft(x)
```

```python
X_Mag = abs(X)
X_ang = np.angle(X)
```

```python
plt.figure()
plt.title('Espectro $X(\omega)$', fontsize=18)
plt.plot(np.linspace(-np.pi, np.pi, len(X_Mag)), np.fft.fftshift(X_Mag))
plt.xlabel('$\omega$')
plt.grid()
plt.show()
```
