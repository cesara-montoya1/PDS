# Ejemplo 1. Analisis de Sistemas LTI.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### $y[n]-0.25y[n-2]=x[n]$. Suponiendo que $y[-1]=15$ y $x[n]=n^2$ 

```python
muestras = 10
# condiciones iniciales ascendente ...,y[-2],y[-1]
inicial  = [15.]
```

```python
m_inicial = len(inicial)
m = muestras + m_inicial
```

```python
# vectores
n  = np.arange(-m_inicial,muestras,1)
xi = np.zeros(m, dtype=float)
yi = np.zeros(m, dtype=float)
```

```python
# añade condiciones iniciales
for k in range(0,m_inicial,1):
    xi[k] = 0
    yi[k] = inicial[k]
```

```python
# calcula los siguientes valores
for k in range(m_inicial,m,1):
    xi[k] = n[k]**2
    yi[k] = 0.25*yi[k-2]+xi[k]
```

```python
tabla = np.concatenate([[n],[xi],[yi]], axis=0)
tabla = np.transpose(tabla)
```

```python
# SALIDA
np.set_printoptions(precision=2)
print('muestras: ', muestras)
print('[[  n,  xi,  yi]]')
print(tabla)

# Gráfica
plt.subplot(211)
plt.stem(n,xi,label='x[n]=n**2')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.axvline(0, linestyle="dashed")
plt.legend()

plt.subplot(212)
plt.stem(n,yi,label='y[n]=0.25y[n-2]+x[n]')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.axvline(0, linestyle="dashed")
plt.legend()
plt.show()
```
