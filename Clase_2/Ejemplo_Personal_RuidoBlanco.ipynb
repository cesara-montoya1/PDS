# Ejemplo personal. Simulando efectos del canal (Ruido blanco)

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Función para añadir ruido blanco
def noise(X: np.ndarray, snr: float):
    # Calcular la potencia en dB de la señal original
    X_avg_p = np.mean(np.power(X, 2))
    X_avg_db = 10 * np.log10(X_avg_p)

    # Calcular la potencia en dB del ruido
    noise_avg_db = X_avg_db - snr
    noise_avg_p = np.power(10, noise_avg_db/10)
    
    # Crear el vector de ruido con distribución normal
    noise = np.random.normal(scale = np.sqrt(noise_avg_p), size=len(X))
    return X + noise
```

```python
# Relación señal a ruido deseada en dB
SNR = 18.0

# Vector de tiempo y señal
t = np.linspace(0, 2*np.pi, 20)
signal = np.sin(t)

# Señal con ruido
signal_with_noise = noise(signal, SNR)
```

```python
# Grafica de las señales con y sin ruido
plt.figure()
plt.plot(t, signal, 'b', label = 'Señal sin ruido')
plt.plot(t, signal_with_noise, 'r', label = 'Señal con ruido')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
```
