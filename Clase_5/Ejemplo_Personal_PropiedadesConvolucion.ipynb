# Ejemplo Personal. Propiedades de la convolución.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
import pds
```

```python
h = [2, 5, 10, 0, 0]
n, f = pds.step(N = 7)
```

## Propiedad conmutativa

$f * h = h * f$

```python
conv = np.convolve(h, f)
conv_conm = np.convolve(f, h)

# Calcular el vector τ
tau = np.arange(-len(f) + 1, len(h), 1)

# Grafica de la convolución
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
pds.plot_discrete(tau, conv, xlabel = r"Retraso ($\tau$)")

plt.subplot(1, 2, 2)
pds.plot_discrete(tau, conv_conm, xlabel = r"Retraso ($\tau$)")

plt.show()
```

## Propiedad asociativa

$(f * g) * h = f * (g * h)$

```python
n_g, g = pds.ramp(N = 7)
```

```python
# (f * g) * h
conv_fg = np.convolve(f, g)
conv_assoc_1 = np.convolve(conv_fg, h)

# f * (g * h)
conv_gh = np.convolve(g, h)
conv_assoc_2 = np.convolve(f, conv_gh)

# Calcular el vector τ
tau_assoc = np.arange(-(len(f) + len(g)) + 2, len(h), 1)

# Grafica de la convolución
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
pds.plot_discrete(tau_assoc, conv_assoc_1, xlabel = r"Retraso ($\tau$)")

plt.subplot(1, 2, 2)
pds.plot_discrete(tau_assoc, conv_assoc_2, xlabel = r"Retraso ($\tau$)")

plt.show()
```

## Propiedad distributiva

$(f + g) * h = f * h + g * h$

```python
# (f + g)
f_plus_g = f + g

# (f + g ) * h
conv_dist_1 = np.convolve(f_plus_g, h)

# f * h
conv_f_h = np.convolve(f, h)

# g * h
conv_g_h = np.convolve(g, h)

# f * h + g * h
conv_dist_2 = conv_f_h + conv_g_h

# Calcular el vector τ
tau_dist = np.arange(-len(f_plus_g) + 1, len(h), 1)

# Grafica de la convolución
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
pds.plot_discrete(tau_dist, conv_dist_1, xlabel = r"Retraso ($\tau$)")

plt.subplot(1, 2, 2)
pds.plot_discrete(tau_dist, conv_dist_2, xlabel = r"Retraso ($\tau$)")

plt.show()
```
